```python
import pandas as pd
import numpy as np
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
feat = pd.read_csv("features.csv")
feat.head()
```

```python
feat.isna().sum()
```

```python
train=pd.read_csv('train.csv')
train.head()
```

```python
train.isna().sum()
```

```python
stores = pd.read_csv('stores.csv')
stores.head()
```

```python
stores.isna().sum()
```

```python
data = train.merge(feat, on=['Store', 'Date'], how='inner').merge(stores, on=['Store'], how='inner')
print(data.shape)
```

```python
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data['MarkDown1'] = data['MarkDown1'].fillna(0)
data['MarkDown2'] = data['MarkDown2'].fillna(0)
data['MarkDown3'] = data['MarkDown3'].fillna(0)
data['MarkDown4'] = data['MarkDown4'].fillna(0)
data['MarkDown5'] = data['MarkDown5'].fillna(0)
```

```python
data.isna().sum()
```

```python
# Convert date to datetime format and show dataset information
data['Date'] =  pd.to_datetime(data['Date'])
data.info()

# Splitting Date and create new columns (Day, Month, and Year)
data["Day"]= pd.DatetimeIndex(data['Date']).day
data['Month'] = pd.DatetimeIndex(data['Date']).month
data['Year'] = pd.DatetimeIndex(data['Date']).year
data
```
